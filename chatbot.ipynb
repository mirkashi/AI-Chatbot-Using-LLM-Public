```python
import streamlit as st
from langchain_openai import ChatOpenAI
from langchain_core.runnables import RunnablePassthrough
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
```

```python
import os

# Set your key directly
os.environ["OPENAI_API_KEY"] = "" 

# Now you can use it
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
# Initialize LLM

llm = ChatOpenAI(
    openai_api_key=os.getenv("OPENAI_API_KEY"),  
    temperature = 0.7,            #this is randomness, that how much generic answers we want to generate
    model_name = "gpt-4o-mini"    #model name
)
```

```python
#Creating Histroy in streamlit

if "history" not in st.session_state:
    st.session_state.history = ChatMessageHistory()   
```

```python
#Creating Langchain Conversation

from langchain_core.runnables.history import RunnableWithMessageHistory

conversation = RunnableWithMessageHistory(
    llm,
    lambda session_id: st.session_state.history,   # correct way
    input_messages_key="input",
    history_messages_key="history",
)
```

```python
#Streamlit UI

st.set_page_config(page_title = "LLM Chatbot", page_icon = "By: Mariyam Siddiqui")
st.title("Langchain Chatbot")

user_input = st.text_input("You: ", key="input")

if user_input:
    response = conversation.predict(input=user_input)
    st.session_state.memory.chat_memory.add_user_message(user_input)
    st.session_state.memory.chat_memory.add_ai_message(user_input)
    st.write(f"**{role}:**{message.content}")
```
